````python
# initialize working environment:

import numpy as np
import pandas as pd
import requests
import requests_cache
import lxml.html as lx

import plotnine
import matplotlib.pyplot as plt

import re
import nltk
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
nltk.download("stopwords")

from sklearn.feature_extraction.text import CountVectorizer

requests_cache.install_cache('mycache',expire_after = 300)
````

````python
base_url = 'https://www.sfchronicle.com'
req = requests.get(base_url)
req.raise_for_status()
req.from_cache
````

````python
def extract_url(base_url):
    
    '''
    
    This function is to get all the urls for the article in a base website (In this case is SF Chronical)
    
    The input of the function should be the base url 
    
    The output will be a dataframe with the titles of the articles  and their urls.
    
    '''
    
#     get the response:
    response = requests.get(base_url)
    response.raise_for_status()

#     check the cache status:
    response.from_cache
    
    result = lx.fromstring(response.text)
    result.make_links_absolute(base_url)
    
#     get the path with links in.
    p = result.xpath('//h2[contains(@class,"headline")]/a|//h3[contains(@class,"headline display-above")]/a | //h4/a')

    
#     Initialize empty lists to save the results:
    links = []
    head = []
    
#     Get all the urls by using for loop:
    for i in range(0,len(p)-1):
        if 'article' in p[i].attrib['href']: 
            title = p[i]
            head.append(title.text)
            link = title.attrib['href']
            
        else:
            continue
        
#         some href in xpath do not consist in the same format: 
#         some are the relative path:

        
        links.append(link)
    links
    head
#     return to a pandas dataframe with articles titles:

    return pd.DataFrame({'Title':head, 'URL':links})
    
````

__Exercise 1.3.__ Use your functions from exercises 1.1 and 1.2 to get data frames of articles for the "Biz+Tech" category as well as two other categories of your choosing (except for "Vault: Archive", "Podcasts", and "In Depth").

Add a column to each that indicates the category, then combine them into one big data frame. Clean up the data, stripping excess whitespace and converting columns to appropriate dtypes.

The `text` column of this data frame will be your corpus for natural language processing in exercise 1.4.

````python
# Grep all the urls by using extract_url funcion:
def getall_articles(base_url):
    '''
    This function is to get all the infomation of the articles under one website.
    This fucntion is actually combinning version of the function in 1.1 and 1.2.
    
    The input of the function should be the link(url) of the base website.
    The output would be a data frame in which all the infomation for each article is listed.
    The category of each article is also in the data frame.
    
    
    '''
    
    
    base_urls = extract_url(base_url)
    base_urls
    base_urls.columns.values
    urls = base_urls.loc[:,'URL']

    # Initialize an empty dataframe to save all the content:
    content = pd.DataFrame(None)
    for url in urls:
        try:
            temp = pd.DataFrame([getinfo(url)])
        except IndexError:
#             print('here is an error')
#             print(url,'\n')
            continue
        content = content.append(temp)
    content = content.reset_index()
    content.drop(labels=['index'],inplace = True, axis = 1)
    
    # get the category
    response = requests.get(base_url)
    response.raise_for_status()
    
    result = lx.fromstring(response.text)
    cat = result.xpath('//body[@id = "home"]')[0].attrib['class']
    content['Category'] = cat
    return content
````

````python
from nltk.corpus import wordnet

def wordnet_pos(tag):
    """Map a Brown POS tag to a WordNet POS tag."""
    
    table = {"N": wordnet.NOUN, "V": wordnet.VERB, "R": wordnet.ADV, "J": wordnet.ADJ}
    
    # Default to a noun.
    return table.get(tag[0], wordnet.NOUN)
````

````python
def get_words(article):
    
    '''
    This function is to get the standardlized word for an article.
    Standarlization means the words will be:
    
    1. In lower case.
    2. Lemmatized.
    3. Without any punctuation.
    
    
    
    '''
    
#     stemmer = nltk.PorterStemmer()
    stopwords = nltk.corpus.stopwords.words('english')
    lemmatizer = nltk.WordNetLemmatizer()
    
    
    
    
    # standarlize the words:
    article = article.lower()
    # Tokenlize the doc:
    article =  nltk.word_tokenize(article)
    

    # Remove stopwords:
    
    words = [w for w in article if w not in stopwords]
    words

    # Lammatize the words:
    
    # [stemmer.stem(w) for w in words]


    words = [w for w in words if w.isalnum()]
    
    pos_tags  = nltk.pos_tag(words)
    pos_tags = [ t for (w,t) in pos_tags ]


    wordnet_tags = [wordnet_pos(tag) for tag in pos_tags]

    
    words_lemmatized = []
    for i in range(0, len(words)-1):
        words_lemmatized.append(lemmatizer.lemmatize(words[i], wordnet_tags[i])) 
    
    return words_lemmatized

````

````python
# Check the frequecy:

    # isalnum() function in Python programming language 
    # checks whether all the characters in a given string is alphanumeric or not.
    
    
def get_fq(words_list):
    '''
    This function is to get the frequncy of the words in one article.
    The output would be an object instead of a list.
    
    '''
    fq = nltk.FreqDist(w for w in words_list if w.isalnum())
    fq
    return fq
    
fq_rs1 = get_fq(rs1)
````
